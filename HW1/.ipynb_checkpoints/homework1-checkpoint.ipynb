# Introduction  

The aim of this report is to perform some analysis on the last 6 year's debit/credit card expenditure data which was obtained from the Central Bank of the Republic of Turkey. 

The following are the datasets that will be used to provide insights:   

1. Total credit/debit card expenditure amount (Thousand TRY)
2. Total number of credit/debit card transactions
3. USD/TRY exchange rate
4. Consumer price index of Turkey

The main goal is to examine the behaviour of the total credit/debit card expenditure amount with the help of other datasets. (All of these data belongs to the Central Bank of the Republic of Turkey.)

Here are the problems that draws the attention after the first look into the data:

To fix these problems some data manipulations have to be done. 

--------

# Data Manipulations  

Data manipulations for each dataset can be found below.  

1. Total credit/debit card expenditure amount (Thousand TRY)

```python
total_expenditure = total_expenditure.rename(columns = {'TP KKHARTUT KT1':'Amount', 'Tarih':'Date'}).dropna()
total_expenditure
```

2. Total number of credit/debit card transactions 

```python
num_transactions = num_transactions.drop(columns=['Unnamed: 2']).dropna()
num_transactions = num_transactions.rename(columns = {'TP KKISLADE KA1':'Number', 'Tarih':'Date'})
num_transactions.drop(num_transactions.tail(2).index,inplace=True)
num_transactions
```

3. USD/TRY exchange rate 

```python
usd_try = usd_try.drop(columns=['Unnamed: 2'])
usd_try = usd_try.rename(columns = {'TP DK USD A YTL':'USDTRY', 'Tarih':'Date'})
usd_try.drop(usd_try.tail(8).index,inplace=True)
usd_try = usd_try.fillna(method='ffill')
usd_try
```

4. Consumer price index of Turkey 

```python
turkey_cpi = turkey_cpi.drop(columns=['Unnamed: 2'])
turkey_cpi = turkey_cpi.rename(columns = {'TP FG J0':'CPI', 'Tarih':'Date'})
turkey_cpi.drop(turkey_cpi.tail(12).index,inplace=True)
turkey_cpi
```

At this moment datasets are ready to be used for the analysis.

# Analysis  

## Number of Transactions Made  

The first question that can be answered in this analysis is that whether there is a positive correlation between the amount spent and the number of transactions. Firstly, these two variables have to be plotted together. To achieve that these two data frames should be merged.

```python
amount_with_transactions = pd.merge(total_expenditure, num_transactions, on='Date')
amount_with_transactions['Amount_Per_Transaction'] =  amount_with_transactions.Amount * 1000 / amount_with_transactions.Number
```

```python
amount_with_transactions
```

There is also a new column called `Amount_Per_Transaction` which represents the average amount spent per transaction.   

Now, the the object is ready to be plotted. 

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(amount_with_transactions.Date, amount_with_transactions.Amount, color='lightblue')
ax.plot(amount_with_transactions.Date, amount_with_transactions.Number, color='darkred')
ax.set_title('Total Expenditure Amount vs # of Transactions')
ax.set_xlabel('Date')
ax.set_ylabel('Total Expenditure (1000 TL)')
plt.show()
```

```python
amount_with_transactions.plot('Date', 'Number', color='darkred')
plt.title('Total Expenditure Amount vs # of Transactions')
plt.xlabel('Date')
plt.ylabel('Transaction')
plt.xticks(rotation = 45)
plt.show()
```
